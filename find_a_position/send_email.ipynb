```python
import pandas as pd
import ipywidgets as widgets

# create input fields for name, email, and university
name_input = widgets.Text(description='Name:')
email_input = widgets.Text(description='Email:')
university_input = widgets.Text(description='University:')

# create a button to add the data to the dataframe
add_button = widgets.Button(description='Add')

# create a dataframe to store the data
df = pd.DataFrame(columns=['Name', 'Email', 'University'])

# define a function to add the data to the dataframe and save it to an Excel file
def add_data(_):
    data = {'Name': name_input.value, 'Email': email_input.value, 'University': university_input.value}
    df.loc[len(df)] = data
    df.to_excel('recipient_info.xlsx', index=False)
    name_input.value = ''
    email_input.value = ''
    university_input.value = ''

# attach the add_data function to the button
add_button.on_click(add_data)

# create a panel with the input fields and button
panel = widgets.VBox([name_input, email_input, university_input, add_button])

# display the panel
display(panel)
```

```python
import pandas as pd
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication

# read the Excel file containing recipient information
df = pd.read_excel('recipient_info.xlsx')

# iterate over each row in the dataframe and send a personalized email to the recipient
for index, row in df.iterrows():

    # create a message object
    msg = MIMEMultipart()
    
    # set the sender, recipient, and subject of the email
    msg['From'] = 'your email'
    msg['To'] = row['Email']
    msg['Cc'] = 'your Cc email'
    msg['Subject'] = 'your subject'
    
    # example text email body
    body = f"Dear Prof. {row['Name']},\n\nI hope this message finds you well. I am interested in pursuing a graduate program in the Department of Computer Science at {row['University']}.\
\n\nSincerely,\
\nMojtaba Mirzadeh"


    # attach the body to the message object
    msg.attach(MIMEText(body))
    
    # attach a PDF file (for example, your CV) to the email
    filename = 'for ex: CV Address'
    with open(filename, 'rb') as f:
        attach = MIMEApplication(f.read(),_subtype='pdf')
        attach.add_header('Content-Disposition','attachment',filename=filename)
        msg.attach(attach)
    
    # authenticate and send the email using your email account information
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login('your email', 'your email password')
    text = msg.as_string()
    server.sendmail('your email', row['Email'], text)
    server.quit()
    
    print(f"Email sent to {row['Name']} at {row['Email']}")

print("All emails have been sent!")
```
